# Importing the required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import io
import plotly.express as ps
```

# Reading the data

```python
from google.colab import files


uploaded = files.upload()
```

```python
df = pd.read_csv(io.BytesIO(uploaded['Unemployment_Rate_upto_11_2020.csv']))
df.head(10)
```

# **DATA ANALYSIS AND VISUALIZATION**:-

# Correlation Matrix:-

```python
corr=df.corr()
fig,ax = plt.subplots(figsize = (5,5))
sns.heatmap(corr,annot=True,ax=ax)
```

## Scatterplot to check the relationship between Estimated Unemployment Rate and Estimated Employed

```python
sns.scatterplot(data =df, x=" Estimated Unemployment Rate (%)", y=" Estimated Employed")
plt.show()
```

## Scatterplot to check the relationship between Estimated Unemployment Rate and Estimated Labour Participation Rate

```python
sns.scatterplot(data=df, x=" Estimated Unemployment Rate (%)", y=" Estimated Labour Participation Rate (%)")
plt.show()
```

# Analyzing Data By Bar Graph

```python
fg = ps.bar(df,x='Region.1',y=' Estimated Unemployment Rate (%)',color = 'Region',
            title='Region Wise Unemployment Rate',template = 'plotly')
fg.show()
```

```python
fg = ps.bar(df,x='Region',y=' Estimated Unemployment Rate (%)',color = 'Region',
            title='State Wise Unemployment Rate',template = 'plotly')
fg.show()
```

```python
fg = ps.scatter(df,x='Region.1',y=' Estimated Unemployment Rate (%)',color = 'Region',
            title='Region Wise Unemployment Rate',template = 'plotly')
fg.show()
```
